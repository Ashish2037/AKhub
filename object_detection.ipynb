```python
import os
import random
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

path = r"C:\Users\Ashish Kumar Singh\Desktop\ShelfImages"
train_file_names = os.listdir(path)

# Ensure train_file_names is not empty before trying to select a random image
if train_file_names:
    image_name = random.choice(train_file_names)
    image_path = os.path.join(path, image_name)
    img = mpimg.imread(image_path)
    plt.imshow(img)
    plt.title("Random Image")
    plt.axis('off')  # To remove axis ticks and labels
    plt.show()
else:
    print("No image files found in the directory.")
```
