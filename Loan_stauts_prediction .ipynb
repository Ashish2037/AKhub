```python
import pandas as pd
```

```python
df=pd.read_csv('loan_data_set.csv')
```

```python
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df=df.drop('Loan_ID',axis=1)
```

```python
df
```

```python
# to fill the NA values
```

```python
gender_mode=df['Gender'].mode()[0]
```

```python
gender_mode
```

```python
df['Gender'].fillna(gender_mode,inplace=True)
```

```python
Married_mode=df['Married'].mode()[0]
```

```python
df['Married'].fillna(Married_mode,inplace=True)
```

```python
Dependents_mode=df['Dependents'].mode()[0]
```

```python
df['Dependents'].fillna(Dependents_mode,inplace=True)
```

```python
Self_Employed_mode=df['Self_Employed'].mode()[0]
```

```python
df['Self_Employed'].fillna(Self_Employed_mode,inplace=True)
```

```python
LoanAmount_mean=df['LoanAmount'].mean()
```

```python
df['LoanAmount'].fillna(LoanAmount_mean,inplace=True)
```

```python
Loan_Amount_Term_mean=df['Loan_Amount_Term'].mean()
```

```python
df['Loan_Amount_Term'].fillna(Loan_Amount_Term_mean,inplace=True)
```

```python
Credit_History_mode=df['Credit_History'].mode()[0]
```

```python
df['Credit_History'].fillna(Credit_History_mode,inplace=True)
```

```python
df.isnull().sum()
```

```python
# to fill the 0's in the co-applicant income 
```

```python
CoapplicantIncome_mean=df['CoapplicantIncome'].mean()
```

```python
df.replace({'CoapplicantIncome':{0:CoapplicantIncome_mean}},inplace=True)
```

```python
df
```

```python
df['Gender'].unique()
```

```python
# dealing with the categorical data
```

```python
df.replace({'Gender':{'Male':1,'Female':0}},inplace=True)
```

```python
df.replace({'Married':{'Yes':1,'No':0}},inplace=True)
```

```python
df['Education'].value_counts()
```

```python
df.replace({'Education':{'Graduate':1,'Not Graduate':0}},inplace=True)
```

```python
df['Self_Employed'].value_counts()
```

```python
df.replace({'Self_Employed':{'Yes':1,'No':0}},inplace=True)
```

```python
df['Property_Area'].value_counts()
```

```python
df.replace({'Property_Area':{'Semiurban':1,'Urban':0,'Rural':2}},inplace=True)
```

```python
df
```

```python
df['Dependents'].unique()
```

```python
df.replace({'Dependents':{'3+':3}},inplace=True)
```

```python
X=df.iloc[:,:11]
X
```

```python
y=df.iloc[:,-1:].values
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
y=le.fit_transform(y)
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=43)
```

```python
X_train
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc=RandomForestClassifier()
```

```python
rfc.fit(X_train,y_train)
```

```python
y_p=rfc.predict(X_test)
```

```python
for i in range(len(y_test)):
    print(y_p[i],y_test[i])
```

```python
from sklearn import metrics
```

```python
X_test_accuracy=metrics.accuracy_score(y_test,y_p)
```
