```python
from urllib.request import urlopen
from bs4 import BeautifulSoup as soup
```

```python
data='https://en.wikipedia.org/wiki/List_of_countries_and_dependencies_by_area'
```

```python
data=urlopen(data)
```

```python
data_html=data.read()
```

```python
data_soup=soup(data_html,'html.parser')
```

```python
head=data_soup.title.string
print(head)
```

```python
tables=data_soup.find_all('table',{'class':'wikitable sortable'})   
```

```python
tables=tables[0]
```

```python
headings=tables.find_all('th')
```

```python
col_title=[]
for i in headings:
    col_title.append(i.text)
```

```python
col_title
```

```python
row_data=tables.find_all('tr')
```

```python
row_data
```

```python
table_row=[]
for rows in row_data:
    cur_row=[]
    r_data=rows.find_all('td')
    for a in r_data:
        cur_row.append(a.text)
    table_row.append(cur_row)
```

```python
import pandas as pd
```

```python
df=pd.DataFrame(table_row,columns=col_title)
```
